## Examples

### example 1

````python
from pyLong.layouts.layout import Layout
from pyLong.layouts.sublayout import Sublayout

layout = Layout()
layout.subdivisions = 3

sublayout = Sublayout()
sublayout.subdivisions = 2

layout.sublayouts = [sublayout]

ax_z, ax_s, subaxs, figure = layout.new_figure()
figure
````
