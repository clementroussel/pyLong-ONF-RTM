## Examples

### example 1

````python
from pyLong.layouts.layout import Layout
from pyLong.layouts.sublayout import Sublayout

layout = Layout()
layout.subdivisions = 2

sublayout = Sublayout()
sublayout.y_axis.min = -25.
sublayout.y_axis.max = +25.
sublayout.y_axis.label = "width [m]"
sublayout.y_axis.intervals = 5
sublayout.y_axis.label_color = "Black"
sublayout.y_axis.value_color = "Gray"
sublayout.y_axis.lower_shift = 5.
sublayout.y_axis.upper_shift = 5.

layout.sublayouts = [sublayout]

ax_z, ax_s, subaxs, figure = layout.new_figure()
figure
````
