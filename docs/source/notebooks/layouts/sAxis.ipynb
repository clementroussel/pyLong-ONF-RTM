## Examples

### example 1

````python
from pyLong.layouts.layout import Layout

layout = Layout()
layout.secondary_axis = True
layout.slope_unit = "%"

layout.s_axis.min_percent = 10.
layout.s_axis.max_percent = 90.
layout.s_axis.label = "slope [m/100m]"
layout.s_axis.intervals_percent = 8
layout.s_axis.decimals_percent = 1
layout.s_axis.label_size = 10.
layout.s_axis.label_color = "Black"
layout.s_axis.value_size = 9.
layout.s_axis.value_color = "Gray"
layout.s_axis.lower_shift_percent = 10.
layout.s_axis.upper_shift_percent = 10.

ax_z, ax_s, subaxs, figure = layout.new_figure()
figure
````
