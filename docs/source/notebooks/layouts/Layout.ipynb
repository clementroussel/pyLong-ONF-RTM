### example 2

````python
from pyLong.layouts.layout import Layout
from pyLong.layouts.sublayout import Sublayout

layout = Layout()
layout.name = "main layout"
layout.width = 21.
layout.height = 21.
layout.secondary_axis = True
layout.slope_unit = "Â°"
layout.as_km = False
layout.subdivisions = 3
layout.hspace = 0.075
layout.pad = 1.

sublayout = Sublayout()
layout.sublayouts = [sublayout]

ax_z, ax_s, subaxs, figure = layout.new_figure()
figure
````
