```python
from pyLong.profiles.zprofile import zProfile
```

```python
zprofile = zProfile()
zprofile.label = "profil brut"
```

```python
from pyLong.layouts.layout import Layout
from pyLong.layouts.sublayout import Sublayout

layout = Layout()

layout.name = "main layout"
layout.width = 21.
layout.height = 21.
layout.secondary_axis = True
layout.slope_unit = "°"
layout.as_km = False
layout.subdivisions = 3
layout.hspace = 0.075
layout.pad = 1.

layout.legend.font_size = 8.
layout.legend.frame = True
layout.legend.opacity = 1

sublayout = Sublayout()
sublayout.legend.font_size = 14
sublayout.legend.font_color = "Red"
sublayout.legend.font_style = "italic"
sublayout.legend.font_weight = "bold"
sublayout.legend.frame = False
layout.sublayouts = [sublayout]

ax_z, ax_s, subaxs, figure = layout.new_figure()

zprofile.plot(ax_z)
zprofile.duplicate().plot(subaxs[0])

#ax_z.legend()
layout.plot_legend(ax_z, subaxs)
figure
```
