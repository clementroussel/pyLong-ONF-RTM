## Paramètres

### Profil en long

````python
# création
zprofile = zProfile()

# importation des valeurs depuis un fichier .txt
zprofile.from_txt("C:/Users/cr26739/Desktop/2022_Dossiers/2022_pyLong_FOP_trajectographie/theorie/profil_contre_pente.txt")

# configuration du style
zprofile.marker_style = 'square'
zprofile.marker_size = 3
````

````python
# tracé
fig, ax = plt.subplots()

ax.set_xlim(-10., 110.)
ax.set_ylim(400., 485.)
ax.set_aspect("equal")

zprofile.plot(ax)

plt.show()
````

````python
# ré-échantillonage du profil
resampled_zprofile = zprofile.resample(dx=2)

# copie du style
resampled_zprofile.copy_style(zprofile)

# titre
resampled_zprofile.title = "profil en long ré-échantillonné"

# légende
resampled_zprofile.label = "profil en long"
````

````python
# tracé du profil ré-échantillonné
fig, ax = plt.subplots()

ax.set_xlim(-10., 110.)
ax.set_ylim(400., 485.)
ax.set_aspect("equal")

resampled_zprofile.plot(ax)

plt.show()
````

## Exploitation des résultats

### Angles d'énergie
```python
>>> mezap.energy_angles

returns:
- energy angles (deg) - type: list
```

````python
for i, beta in enumerate(mezap.energy_angles):
    if i < 5:
        display(Markdown(
            rf"$\beta_{i} = {round(beta, 2)}$ deg"))
````

### Aires des trapèzes élémentaires
```python
>>> mezap.areas

returns:
- areas (m^2) - type: list
```

````python
for i, area in enumerate(mezap.areas):
    if i < 5:
        display(Markdown(
            rf"$A_{i} = {round(area, 2)}$ $m^2$"))
````

### Aires cumulées
```python
>>> mezap.cumulated_ares

returns:
- cumulated areas (m^2) - type: list
```

````python
for i, area in enumerate(mezap.cumulated_areas):
    if i < 5:
        display(Markdown(
            rf"$A^c_{i} = {round(area, 2)}$ $m^2$"))
````

### Aires normalisées
```python
>>> mezap.normalized_ares

returns:
- normalized areas (m^2) - type: list
```

````python
for i, area in enumerate(mezap.normalized_areas):
    if i < 5:
        display(Markdown(
            rf"$A^n_{i} = {round(area, 2)}$ $m^2$"))
````

## Chargement de la base de données

````python
limits = np.loadtxt("../../../notebooks/toolbox/mezap/limits.txt")

df = pd.read_csv("../../../notebooks/toolbox//mezap/sample.txt", delimiter='\t', decimal=',')
sample = np.array(df)
````

````python
# tracé de la base de données
fig, ax = plt.subplots()

ax.set_xlim(0., 1.)
ax.set_ylim(0., 90.)

ax.set_xlabel(r"$A^n$")
ax.set_ylabel(r"$\beta$ (deg)")

ax.grid(True, linestyle="--", alpha=0.3, which="both")
ax.minorticks_on()

ax.plot(limits[:,0], limits[:,3], 'Red', label="proba. d'atteinte forte", zorder=1)
ax.plot(limits[:,0], limits[:,2], 'Orange', label="proba. d'atteinte moyenne", zorder=1)
ax.plot(limits[:,0], limits[:,1], 'Green', label="proba. d'atteinte faible", zorder=1)

ax.scatter(sample[:, 0], sample[:, 1], s=0.5, color='blue', label="échantillon")

ax.legend(loc="best", fontsize=8)

plt.show()
````

## Animation

````python
def f(i):
    profile = resampled_zprofile.duplicate()

    fig = plt.figure(constrained_layout=True, figsize=(1.4*29.7/2.54, 1.4*21/2.54))
    gs = GridSpec(4, 2, figure=fig)

    ax_profile = fig.add_subplot(gs[0:2, 0])
    ax_mezap = fig.add_subplot(gs[0:2, 1])
    ax_angles = fig.add_subplot(gs[2, 0], sharex=ax_profile)
    ax_areas = fig.add_subplot(gs[3, 0], sharex=ax_profile)
    ax_summary = fig.add_subplot(gs[2:, 1])
    ax_summary.axis('off')


    ax_profile.set_xlim(min(profile.x)-0.05*(max(profile.x)-min(profile.x)), max(profile.x)+0.05*(max(profile.x)-min(profile.x)))
    ax_profile.set_ylim(min(profile.z)-0.05*(max(profile.z)-min(profile.z)), max(profile.z)+0.05*(max(profile.z)-min(profile.z)))
    ax_profile.set_xlabel("x (m)")
    ax_profile.set_ylabel("z (m)")
    ax_profile.minorticks_on()

    ax_mezap.set_xlim(min(mezap.normalized_areas[1:])-0.05*(max(mezap.normalized_areas[1:])-min(mezap.normalized_areas[1:])),
                      max(mezap.normalized_areas[1:])+0.05*(max(mezap.normalized_areas[1:])-min(mezap.normalized_areas[1:])))
    ax_mezap.set_ylim(min(mezap.energy_angles[1:])-0.05*(max(mezap.energy_angles[1:])-min(mezap.energy_angles[1:])),
                      max(mezap.energy_angles[1:])+0.05*(max(mezap.energy_angles[1:])-min(mezap.energy_angles[1:])))
    ax_mezap.set_xlabel(r"$A^n$")
    ax_mezap.set_ylabel(r"$\beta$ (deg)")
    ax_mezap.grid(True, linestyle="--", alpha=0.3, which="both")
    ax_mezap.minorticks_on()

    ax_angles.set_ylim(min(mezap.energy_angles[1:])-0.05*(max(mezap.energy_angles[1:])-min(mezap.energy_angles[1:])),
                       max(mezap.energy_angles[1:])+0.05*(max(mezap.energy_angles[1:])-min(mezap.energy_angles[1:])))
    ax_angles.set_xlabel("x (m)")
    ax_angles.set_ylabel(r"$\beta$ (deg)")
    ax_angles.grid(True, linestyle="--", alpha=0.3, which="both")
    ax_angles.minorticks_on()

    ax_areas.set_ylim(min(mezap.normalized_areas[1:])-0.05*(max(mezap.normalized_areas[1:])-min(mezap.normalized_areas[1:])),
                      max(mezap.normalized_areas[1:])+0.05*(max(mezap.normalized_areas[1:])-min(mezap.normalized_areas[1:])))
    ax_areas.set_xlabel("x (m)")
    ax_areas.set_ylabel(r"$A^n$")
    ax_areas.grid(True, linestyle="--", alpha=0.3, which="both")
    ax_areas.minorticks_on()


    # ligne d'énergie
    ax_profile.plot([mezap.zprofile.x[0]+mezap.start[0], mezap.zprofile.x[i]+mezap.start[0]], [mezap.start[1], mezap.zprofile.z[i]], "--", zorder=1, label="ligne d'énergie")

    # aire cumulée
    xz = np.zeros((len(mezap.zprofile.xz[:i+1])+1, 2))
    for i, (x, z) in enumerate(mezap.zprofile.xz[:i+1]):
        xz[i, 0], xz[i, 1] = x+mezap.start[0], z

    xz[-1,0] = mezap.start[0]
    xz[-1,1] = mezap.zprofile.z[i]

    cumulated_area = Polygon(xz, closed=True)
    cumulated_area.set_facecolor("None")
    cumulated_area.set_edgecolor("Blue")
    cumulated_area.set_hatch(2*"/")
    cumulated_area.set_label("aire cumulée $A^c$")

    ax_profile.add_artist(cumulated_area)
    
    ax_profile.plot([mezap.zprofile.x[i]+mezap.start[0]], [mezap.zprofile.z[i]], 'ms')

    # aire de normalisation
    square = Rectangle((mezap.start[0], mezap.zprofile.z[i]),
                        mezap.zprofile.z[0] - mezap.zprofile.z[i],
                        mezap.zprofile.z[0] - mezap.zprofile.z[i])
    square.set_edgecolor("Blue")
    square.set_facecolor("Blue")
    square.set_alpha(0.2)
    square.set_label(r"$(z - z_{start})^2$")

    ax_profile.add_artist(square)

    # profil en long
    profile.plot(ax_profile)

    ax_profile.legend(loc="best", fontsize=8)

    line_mezap, = ax_mezap.plot(mezap.normalized_areas[1:], mezap.energy_angles[1:])
    line_mezap.set_label("profil normalisé")
    line_mezap.set_color("Black")
    line_mezap.set_marker("s")
    line_mezap.set_markersize(3)
    line_mezap.set_zorder(2)

    ax_mezap.plot(limits[:,0], limits[:,3], 'Red', label="proba. d'atteinte forte", zorder=1)
    ax_mezap.plot(limits[:,0], limits[:,2], 'Orange', label="proba. d'atteinte moyenne", zorder=1)
    ax_mezap.plot(limits[:,0], limits[:,1], 'Green', label="proba. d'atteinte faible", zorder=1)

    ax_mezap.scatter(sample[:, 0], sample[:, 1], s=0.5, color='blue', label="échantillon")

    ax_mezap.plot([mezap.normalized_areas[i]], [mezap.energy_angles[i]], 'ms', zorder=2)
    
    ax_mezap.legend(loc="best", fontsize=8)

    line_angle, = ax_angles.plot(np.array(mezap.zprofile.x[1:])+mezap.start[0], mezap.energy_angles[1:])
    line_angle.set_label(r"angle d'énergie $\beta$")
    line_angle.set_color("Black")
    line_angle.set_marker("s")
    line_angle.set_markersize(3)
    
    ax_angles.plot([mezap.zprofile.x[i]+mezap.start[0]], [mezap.energy_angles[i]], 'ms')

    ax_angles.legend(loc="best", fontsize=8)

    line_area, = ax_areas.plot(np.array(mezap.zprofile.x[1:])+mezap.start[0], mezap.normalized_areas[1:])
    line_area.set_label(r"aire normalisée $A^n$")
    line_area.set_color("Black")
    line_area.set_marker("s")
    line_area.set_markersize(3)
    
    ax_areas.plot([mezap.zprofile.x[i]+mezap.start[0]], [mezap.normalized_areas[i]], 'ms')

    ax_areas.legend(loc="best", fontsize=8)
    
    ax_summary.text(0, 0.95, "pyLong ONF-RTM")
    ax_summary.text(0, 0.85,"Calcul de lignes d’énergie à partir de l’aire normalisée et d’échantillon de valeurs relevées\n(Méthode MEZAP, rapport BRGM RP-66589-FR).")
    ax_summary.text(0, 0.75,"Les échantillons de valeurs sont issus d’une base de données collectées par l’IRSTEA et le\nBRGM (dont des données RTM).")
    ax_summary.text(0, 0.65,"Les valeurs de probabilité d’atteinte forte, moyenne et faible sont des valeurs proposées\ndans le cadre de la méthode MEZAP par analyse statistique de l’échantillon.")
    
    ax_summary.text(0.45, 0.55, r"$x_{start} =$" + rf"${round(mezap.start[0], 3)}$ m")
    ax_summary.text(0.45, 0.5, r"$z_{start} =$" + rf"${round(mezap.start[1], 3)}$ m")
    
    ax_summary.text(0.45, 0.4, rf"$x = {round(mezap.zprofile.x[i]+mezap.start[0], 3)}$ m")
    ax_summary.text(0.45, 0.35, rf"$z = {round(mezap.zprofile.z[i], 3)}$ m")
    ax_summary.text(0.45, 0.3, r"$distance =$" + rf"${round(mezap.distances[i], 3)}$ m")
    
    ax_summary.text(0.45, 0.2, rf"$\beta = {round(mezap.energy_angles[i], 2)}$ deg")
    ax_summary.text(0.45, 0.15, rf"$A^t = {round(mezap.areas[i], 2)}$ $m^2$")
    ax_summary.text(0.45, 0.1, rf"$A^c = {round(mezap.cumulated_areas[i], 2)}$ $m^2$")
    ax_summary.text(0.45, 0.05, r"$(z - z_{start})^2 =" + rf"{round(mezap.denominators[i], 2)}$ $m^2$")
    ax_summary.text(0.45, 0., rf"$A^n = {round(mezap.normalized_areas[i], 2)}$")
    
    plt.tight_layout()
    
    plt.savefig(f"C:/Users/cr26739/Desktop/2022_Dossiers/2022_pyLong_FOP_trajectographie/png_contre_pente/figure_{i}")

    plt.show()
````

````python
r = ipywidgets.interact(f, i=ipywidgets.BoundedIntText(value=1, min=1, max=len(mezap.areas)-1, step=1, description="Vertice n°"))
````
