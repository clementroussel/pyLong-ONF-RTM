```python
import random

from matplotlib import pyplot as plt
import numpy as np

from pyLong.dictionnaries.colors import colors as html_colors
from pyLong.toolbox.platrock2d_shape_viewer import PlatRock2DShapeViewer
from pyLong.profiles.zprofile import zProfile
```

# Visionneuse **PlatRock 2D Shape**

## Création d'une nouvelle visionneuse

```python
viewer = PlatRock2DShapeViewer()
```

## Importation du fichier de résultats *full_output.hdf5*

```python
#--- à configurer par l'utilisateur ---#
filename = r"full_output.hdf5"
#--------------------------------------#

filename = filename.replace("\\", "/")

if viewer.from_hdf5(filename):
    print("L'importation a réussi.")
else:
    print("Echec de l'importation.")
```

## Nombre de blocs simulés

```python
N = viewer.rocks_number()

print(f"La simulation contient {N} blocs.")
```

## Affichage des trajectoires

### Création du profil en long du terrain naturel

```python
#--- à configurer par l'utilisateur ---#
filename = r"terrain.csv"
#--------------------------------------#

filename = filename.replace("\\", "/")

terrain = zProfile()

if terrain.from_txt(filename):
    print("L'importation a réussi.")
else:
    print("Echec de l'importation.")
```

### Modification du style graphique du terrain naturel

```python
terrain.label = "terrain naturel"
terrain.line_color = "Green"
terrain.line_thickness = 1.5
terrain.order = 2
```

### Générations des trajectoires des blocs

```python
trajectories = []

for i in range(N):
    trajectory = viewer.get_trajectory(i, dx=1.)
    
    if trajectory is not None:
        trajectories.append(trajectory)
```

### Modification du style graphique des trajectoires

```python
for trajectory in trajectories:
    trajectory.line_color = random.choice(list(html_colors.keys())) # attribution d'une couleur aléatoire
    trajectory.line_thickness = 0.8
```

### Tracé

```python
# création d'une nouvelle figure
fig, ax = plt.subplots(figsize=(29.7/2.54, 21/2.54))

# configuration des axes
ax.set_xlabel("x [m]")
ax.set_ylabel("z [m]")

#--- à configurer par l'utilisateur ---#
ax.set_xlim((-5, 190))
ax.set_ylim((245, 350))
#--------------------------------------#

# ratio X/Z = 1 (aucune déformation d'échelle)
ax.set_aspect('equal')

# tracé du terrain naturel
terrain.plot(ax) 

# tracés des trajectoires
for trajectory in trajectories:
    trajectory.plot(ax)

# affichage de la légende
ax.legend(loc='best')

# affichage de la figure
plt.show()
```

## Analyse des distances de départ

```python
start_distances = viewer.x_starts()

print(f"Les {len(start_distances)} distances de départ se répartissent tel que:")
print(f"\t - la distance moyenne vaut: {np.round(np.mean(start_distances), 2)} m")
print(f"\t - la distance médianne vaut: {np.round(np.median(start_distances), 2)} m")
print(f"\t - l'écart-type vaut: {np.round(np.std(start_distances), 2)} m")
print(f"\t - la distance minimale vaut: {np.round(np.min(start_distances), 2)} m")
print(f"\t - la distance maximale vaut: {np.round(np.max(start_distances), 2)} m")
```

## Analyse des distances d'arrivée

```python
end_distances = viewer.x_ends()

print(f"Les {len(end_distances)} distances d'arrivée se répartissent tel que:")
print(f"\t - la distance moyenne vaut: {np.round(np.mean(end_distances), 2)} m")
print(f"\t - la distance médianne vaut: {np.round(np.median(end_distances), 2)} m")
print(f"\t - l'écart-type vaut: {np.round(np.std(end_distances), 2)} m")
print(f"\t - la distance minimale vaut: {np.round(np.min(end_distances), 2)} m")
print(f"\t - la distance maximale vaut: {np.round(np.max(end_distances), 2)} m")
```

## Analyses statistiques pour une distance fixée

### Hauteurs au-dessus du terrain natuel

```python
#--- à configurer par l'utilisateur ---#
distance = 110.
#--------------------------------------#

# hag: "heights above ground"
hag = []

for trajectory in trajectories:
    z_traj = trajectory.interpolate(distance)
    z_tn = terrain.interpolate(distance)
    
    if not (z_traj is None or z_tn is None):
        if z_traj - z_tn >= 0:
            hag.append(z_traj - z_tn)
        else:
            hag.append(0.)

print(f"Sur les {len(trajectories)} trajectoires:")
print(f"\t - {len(hag)} (soit {np.round(100*len(hag)/len(trajectories), 2)}%) atteignent la distance {distance} m")
print(f"\t - la hauteur moyenne vaut: {np.round(np.mean(hag), 2)} m")
print(f"\t - la hauteur médianne vaut: {np.round(np.median(hag), 2)} m")
print(f"\t - l'écart-type vaut: {np.round(np.std(hag), 2)} m")
print(f"\t - la hauteur minimale vaut: {np.round(np.min(hag), 2)} m")
print(f"\t - la hauteur maximale vaut: {np.round(np.max(hag), 2)} m")
```
